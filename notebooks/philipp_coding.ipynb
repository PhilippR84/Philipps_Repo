```python
Fruechte = ["Apfel", "Banane", "Birne"]
for x in Fruechte:
  if x == "Banane":
    continue
  print(x)
```

For-Schleife 

Tabelle Fruechte wird erstellt. 
Inhalt der Tabelle ist "Apfel", "Banane" und "Birne"
Mit einer For-Schleife wird jeder Eintrag in der Tabelle durchlaufen.
x: Momentane Stelle in der Tabelle 

Apfel = Stelle 0 in der Tabelle
Bannae = 1
Birne = 2

If-Abfrage 
Ist X gleich "Banane"
Wenn ja:
Mach mit dem nächsten Wert in der Tabelle weiter
Wenn nein:
print(x) : Gebe X aus.



```python
Fruechte = ["Apfel", "Banane", "Birne"]
for x in Fruechte:
  if x == "Banane":
    break
  print(x)
```

Gleiche For-Schleife wie oben.
Unterschied: 
break: Geht aus der Schleife.

Schleife durch den Range.
Wenn die Schleife fertig ist wird "Bin Fertig!" ausgegeben

```python
#Nested Loops:

Tabelle1 = ["Apfel", "Baum", "Haus"]
Tabelle2 = ["Maus", "Wasser", "Baum"]

for x in Tabelle1:
    for y in Tabelle2:
        if x == y:
            print("Übereinstimmung, das Wort ist: " + x)
        else:
            print("Keine Übereinstimmung " + x + " " + y)
```
